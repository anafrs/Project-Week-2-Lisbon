```python
import pandas as pd
```

```python
agenda = pd.read_json("agenda_cultural.json")
```

```python
agenda.columns
```

```python
agenda_lx = agenda[["ID","categories_name_list", "title", "venue", "occurences", "LastDate", "price_cat" ]]
agenda_lx.head()
```

```python
agenda_lx["LastDate"].isnull().sum()
```

```python
agenda_lx["LastDate"] = pd.to_datetime(agenda_lx["LastDate"])
```

```python
agenda_lx["Month"] = agenda_lx["LastDate"].dt.month
```

```python
agenda_lx.head()
```

```python
agenda_lx["Quarter"] = agenda_lx["LastDate"].dt.quarter
```

```python
agenda_lx.head()
```

### What is the month and quarter with the higher number of events?

```python
# the months with more cultural offer are september, November and December.
# August is the month with the lowest cultural offer
agenda_lx.groupby("Month").count()
```

```python
agenda_lx.groupby("Quarter").count()

# quarter 4 has a lot more offer
```
