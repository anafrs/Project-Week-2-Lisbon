## Getting the venue name

Do a Dictionary comprehension list that gets the value of the last key in venue

```python
venue = agenda["venue"]

def get_venue(x):
    if type(x) == dict:
        # Return the first key
        return list(x.keys())[0]
    
    return "---"

            
agendalx["Venue Name"] = agendalx["venue"].apply(get_venue)
agendalx.head()
```

```python
venue_occur = agendalx['Venue Name'].value_counts()
venue_occur
```
