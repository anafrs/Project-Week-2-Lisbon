```python
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
agenda = pd.read_json("agenda_cultural.json")
agenda.columns
```

```python
agendalx = agenda[["ID", "categories_name_list", "title", "venue", "occurences", "LastDate", "price_cat"]]
agendalx.head(5)
```

### Clean venue  and create column


```python
venue = agenda["venue"]

def get_venue(x):
    if type(x) == dict:
        # Return the first key
        return list(x.keys())[0]
    
    return "---"

            
agendalx["Venue Name"] = agendalx["venue"].apply(get_venue)
agendalx.head()
```

### Clean date and create column 


```python
# no null values
agendalx["LastDate"].isnull().sum()
# to datetime
agendalx["LastDate"] = pd.to_datetime(agendalx["LastDate"])
#month column
agendalx["Month"] = agendalx["LastDate"].dt.month
#quarter column
agendalx["Quarter"] = agendalx["LastDate"].dt.quarter
```

### Clean type of event and create event

```python
def get_event(x):
    if type(x) == dict:
        # Return the first key
        return list(x.keys())[0]
    return "---"
agendalx["categories_name_list"] = agendalx["categories_name_list"].apply(get_event)
```

### Relation between top 20 venues and season

```python
top20_venues = ["centro-cultural-de-belem", "fundacao-calouste-gulbenkian", "sao-luiz-teatro-municipal", "cinemateca", "museu-do-oriente", "coliseu-dos-recreios", "teatro-nacional-dona-maria-ii", "culturgest", "cineteatro-capitolio-teatro-raul-solnado", "altice-arena", "musicbox",
"rua-das-gaivotas-6", "museu-da-marioneta", "maat-museu-de-arte-arquitetura-e-tecnologia", "teatro-tivoli-bbva", "lu-ca-teatro-luis-de-camoes", "casino-de-lisboa", "cinema-sao-jorge", "museu-do-dinheiro", "teatro-do-bairro", "praca-de-toiros-do-campo-pequeno"]
quarters =[1, 2, 3, 4]

new_data = []

# loop over each venue

# for each venue, loop over each quarter

# filter the rows that match the venue name and the quarter

# count that
grouped_venues = agendalx.loc[:,["Venue Name", "Quarter"]]
# print(grouped_venues)


for venue in top20_venues:
    d = { "name": venue }
    for q in quarters:
        count = grouped_venues[grouped_venues["Venue Name"].eq(venue) & grouped_venues["Quarter"].eq(q)].shape[0]
        d["q" + str(q)] = count
        
    new_data.append(d)

new_data_table = pd.DataFrame(new_data)
new_data_table
```
